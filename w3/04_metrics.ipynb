```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, precision_recall_curve

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def save_answer(name, value):
    with open('data/statement-metrics_' + name + '.txt', 'w', encoding='utf-8') as f:
        f.write(value)
```

```python
df = pd.read_csv('data/classification.csv')
```

```python
TP = np.sum(np.where((df.pred==1)&(df.true==1), 1, 0))
FP = np.sum(np.where((df.pred==1)&(df.true==0), 1, 0))
FN = np.sum(np.where((df.pred==0)&(df.true==1), 1, 0))
TN = np.sum(np.where((df.pred==0)&(df.true==0), 1, 0))
```

```python
answer = ' '.join([str(el) for el in [TP, FP, FN , TN]])

answer
```

```python
save_answer('1', answer)
```

```python
accuracy = accuracy_score(df.true, df.pred)
precision = precision_score(df.true, df.pred)
recall = recall_score(df.true, df.pred)
f1 = f1_score(df.true, df.pred)
```

```python
answer = ' '.join([str(round(el, 2)) for el in [accuracy, precision, recall , f1]])

answer
```

```python
save_answer('2', answer)
```
