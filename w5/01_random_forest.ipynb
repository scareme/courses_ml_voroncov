```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import KFold, cross_val_score

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def save_answer(name, value):
    with open('data/statement-forest_' + name + '.txt', 'w', encoding='utf-8') as f:
        f.write(value)
```

```python
df = pd.read_csv('data/abalone.csv')
```

```python
df.Sex.unique()
```

```python
df['Sex'] = df.Sex.map({'M': 1, 'F':-1, 'I':0})
```

```python
X = df.drop('Rings', axis=1)
y = df.Rings
```

```python
kf = KFold(n_splits=5, random_state=1, shuffle=True)
```

```python
answer = []
for n_estimators in range(1, 51):
    print(n_estimators)
    estimator = RandomForestRegressor(random_state=1, n_estimators=n_estimators)
    model = cross_val_score(estimator, X, y, cv=kf, scoring='r2')
    answer.append(model.mean())
```

```python
plt.plot(range(1, 51), answer)
plt.plot([0, 51], [0.52, 0.52])
```

```python
for x, y in zip(range(1, 51), answer):
    if y>0.52: print(x)
```
