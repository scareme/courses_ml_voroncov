```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def save_answer(name, value):
    with open('data/statement-pca_' + name + '.txt', 'w', encoding='utf-8') as f:
        f.write(value)
```

```python
df = pd.read_csv('data/close_prices.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
X = df.drop('date', axis=1)
```

```python
pca = PCA(n_components = 10)
pca.fit(X)
```

```python
pca.explained_variance_ratio_ 
```

```python
sums = []
for n in range(1, len(pca.explained_variance_ratio_)+1):
    sum_i = np.sum(pca.explained_variance_ratio_[:n])
    sums.append(sum_i)
```

```python
sums
```

```python
save_answer('1', '4')
```

```python
X_pca = pca.transform(X)
```

```python
x_pca = X_pca[:,0]
```

```python
djia = pd.read_csv('data/djia_index.csv')
```

```python
djia = djia['^DJI'].values
```

```python
djia.shape, x_pca.shape
```

```python
np.corrcoef(x_pca, djia)
```

```python
answer = np.corrcoef(x_pca, djia)[0,1]
answer = str(round(answer, 2))

answer
```

```python
save_answer('2', answer)
```

```python
answer = pd.Series(np.abs(pca.components_[0]), index=df.drop('date', axis=1).columns).sort_values(ascending=False).head(1).index
answer = answer[0]

answer
```

```python
save_answer('3', answer)
```
