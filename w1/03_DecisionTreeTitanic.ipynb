```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
def save_answer(name, value):
    with open('data/statement-importance_' + name + '.txt', 'w') as f:
        f.write(value)
```

```python
sex_map = {'male': 1, 'female': 0}
```

```python
df = pd.read_csv('data/titanic2.csv')
```

```python
df.head()
```

```python
df.Sex = df.Sex.map(sex_map)
```

```python
X = df[['Age', 'Sex', 'Fare', 'Pclass']]
X = X.dropna()
y = df['Survived'][X.index.values]
```

```python
clf = DecisionTreeClassifier(random_state=241)
clf.fit(X.values, y.values)
```

Вычислите важности признаков и найдите два признака с наибольшей важностью. Их названия будут ответами для данной задачи (в качестве ответа укажите названия признаков через запятую или пробел, порядок не важен)

```python
clf.feature_importances_
```

```python
' '.join(pd.Series(clf.feature_importances_, index=X.columns).nlargest(2).index)
```

```python
save_answer('1', ' '.join(pd.Series(clf.feature_importances_, index=X.columns).nlargest(2).index))
```
