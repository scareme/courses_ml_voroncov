```python
import pandas as pd
import numpy as np

from sklearn.metrics import accuracy_score
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def save_answer(name, value):
    with open('data/statement-linear_' + name + '.txt', 'w', encoding='utf-8') as f:
        f.write(value)
```

```python
df_train = pd.read_csv('data/perceptron-train.csv', header=None)
df_test = pd.read_csv('data/perceptron-test.csv', header=None)
```

```python
X_train = df_train[[1, 2]]
y_train = df_train[0]

X_test = df_test[[1, 2]]
y_test = df_test[0]
```

```python
estimator = Perceptron(random_state=241)
estimator.fit(X_train, y_train)
acc1 = accuracy_score(y_test, estimator.predict(X_test))

acc1
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

estimator = Perceptron(random_state=241)
estimator.fit(X_train, y_train)
acc2 = accuracy_score(y_test, estimator.predict(X_test))

acc2
```

```python
save_answer('1', str(round(acc2-acc1, 3)))

round(acc2-acc1, 3)
```
